## **Agora vamos separar esse DataFrame para organizar os dados**

```python
ibge_sp_1 = pd.DataFrame(ibge_sp[['Nome_do_subdistrito', 'Domicilios', 'Residentes']])
```

```python
ibge_sp_1
```

```python
grupo_1 = ibge_sp_1.groupby('Nome_do_subdistrito').sum()
```

```python
grupo_1['Media_de_residentes_por_domicilio'] = grupo_1['Residentes'] / grupo_1['Domicilios']
```

```python
grupo_1 = grupo_1.round(2)
```

**Nós já criamos um DataFrame apenas com os dados que precisavam ser somados e agora sabemos o número total de domicílios e de residentes de cada subdistrito ou bairro, além da média de residentes por domicílio em cada bairro!!!

Nota-se que os dados de renda são do tipo object e os números estão com virgula na casa decimal, o que o python não entende. Então vamos trocar a vírgula por ponto e trocar o tipo das colunas

```python
ibge_sp.replace({',':'.'}, regex = True, inplace = True)
```

```python
ibge_sp
```

Pronto! Já trocamos a , pelo . Agora vamos trocar o tipo de object para float

```python
ibge_sp[['Media_de_residentes_por_domicilio', 'Renda_responsaveis_por_domicilio', 'Renda_resp_com_rendimento']] = ibge_sp[['Media_de_residentes_por_domicilio', 'Renda_responsaveis_por_domicilio', 'Renda_resp_com_rendimento']].apply(pd.to_numeric)
```

```python
ibge_sp.info()
```

Maravilha, convertemos as três colunas ao mesmo tempo! Agora vamos trabalhar!!

Pronto, os únicos dados nulos agora são verdadeiros, referente a domicílios que realmente não tem renda. 

Agora que já temos as colunas na forma correta, precisamos calcular a renda por bairro. No caso da renda, não podemos somar a renda de todos os domicílios ou residentes do bairro, mas sim fazer a média deles

```python
ibge_sp2 = pd.DataFrame(ibge_sp[['Nome_do_subdistrito', 'Renda_responsaveis_por_domicilio', 'Renda_resp_com_rendimento']])
ibge_sp2.head()
```

Agora nós temos um DataFrame com o nome do bairro, a renda dos responsáveis por domicílio e a renda dos responsáveis por domicílio que possuem renda. Agora vamos tirar a média desses dados 

```python
ibge_sp2.groupby('Nome_do_subdistrito').mean().round(2)
```

Maravilha! Vamos chamar esse grupo de grupo_2 e depois juntar tudo!!!

```python
grupo_2 = ibge_sp2.groupby('Nome_do_subdistrito').mean().round(2)
```

## Agora vamos começar a descobrir algumas coisas

Bairros com renda da pessoa responsávelo pelo domicílio menor que 1000 reais

```python
menor_1000 = dados.query('Renda_responsaveis_por_domicilio < 1000')
menor_1000
```

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
bx = sns.barplot(x = 'Bairro', y = 'Renda_responsaveis_por_domicilio', data = menor_1000)
bx.figure.set_size_inches(12,6)
bx.set_title('Renda média da pessoa responsável pelo domicílio por bairro em São Paulo', loc = 'left', fontsize = 18)
bx.set_xlabel('Bairros cuja renda média da pessoa responsável é é menor de 1000 reais', fontsize = 14)
bx.set_ylabel('Renda em reais (R$)', fontsize = 14)
bx.tick_params(axis='x', rotation = 45)
bx = bx
```

Agora os bairros mais ricos, como será que é a renda?

```python
maior_5000 = dados.query('Renda_responsaveis_por_domicilio > 5000')
maior_5000
```

```python

sns.set_palette('Accent')
sns.set_style('darkgrid')
bx = sns.barplot(x = 'Bairro', y = 'Renda_responsaveis_por_domicilio', data = maior_5000)
bx.figure.set_size_inches(12,6)
bx.set_title('Renda média da pessoa responsável pelo domicílio por bairro em São Paulo', loc = 'left', fontsize = 18)
bx.set_xlabel('Bairros cuja renda média da pessoa responsável é maior de 10000 reais', fontsize = 14)
bx.set_ylabel('Renda em reais (R$)', fontsize = 14)
bx.tick_params(axis='x', rotation = 45)
bx = bx
```

Agora vamos comparar os dois gráficos

```python
comparar = dados.query('Renda_responsaveis_por_domicilio > 5000 | Renda_responsaveis_por_domicilio < 1000')
comparar.sort_values('Renda_responsaveis_por_domicilio', inplace = True)
```

```python
sns.set_palette('Accent')
sns.set_style('darkgrid')
bx = sns.barplot(x = 'Bairro', y = 'Renda_responsaveis_por_domicilio', data = comparar)
bx.figure.set_size_inches(12,6)
bx.set_title('Renda média da pessoa responsável pelo domicílio por bairro em São Paulo', loc = 'left', fontsize = 18)
bx.set_xlabel('Bairros cuja renda média da pessoa responsável é menos de 1000 e maior de 5000 reais', fontsize = 14)
bx.set_ylabel('Renda em reais (R$)', fontsize = 14)
bx.tick_params(axis='x', rotation = 90)
bx = bx
```
